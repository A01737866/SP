```python
#Cargar desde un archivo .csv sin indice
df = pd.read_csv('SP_1.csv')
df
```

```python
df['¿Cuál es su grado de satisfacción con las soluciones tecnológicas que ha implantado en su tienda?']= df['¿Cuál es su grado de satisfacción con las soluciones tecnológicas que ha implantado en su tienda?'].astype(float)
```

```python
df['¿Cuál es su grado de satisfacción con las soluciones tecnológicas que ha implantado en su tienda?']=df['¿Cuál es su grado de satisfacción con las soluciones tecnológicas que ha implantado en su tienda?'].str.replace('Muy satisfecho', '5')
df['¿Cuál es su grado de satisfacción con las soluciones tecnológicas que ha implantado en su tienda?']=df['¿Cuál es su grado de satisfacción con las soluciones tecnológicas que ha implantado en su tienda?'].str.replace('Algo satisfecho', '4')
df['¿Cuál es su grado de satisfacción con las soluciones tecnológicas que ha implantado en su tienda?']=df['¿Cuál es su grado de satisfacción con las soluciones tecnológicas que ha implantado en su tienda?'].str.replace('Muy insatisfecho', '1')
df['¿Cuál es su grado de satisfacción con las soluciones tecnológicas que ha implantado en su tienda?']=df['¿Cuál es su grado de satisfacción con las soluciones tecnológicas que ha implantado en su tienda?'].str.replace('Algo insatisfecho', '2')
```

```python
df['Antiguedad del negocio']= df['Antiguedad del negocio'].astype(float)
```

```python
df['Antiguedad del negocio']= df['Antiguedad del negocio'].str.replace('Entre 0 y 6 meses', '1')
df['Antiguedad del negocio']= df['Antiguedad del negocio'].str.replace('Entre 6 meses y un año', '2')
df['Antiguedad del negocio']= df['Antiguedad del negocio'].str.replace('Entre 1 y 3 años', '3')
df['Antiguedad del negocio']= df['Antiguedad del negocio'].str.replace('Entre 3 y 5 años', '4')
df['Antiguedad del negocio']= df['Antiguedad del negocio'].str.replace('Más de 5 años', '5')
```

```python
df['Rol de la persona entrevistada']= df['Rol de la persona entrevistada'].astype(float)
```

```python
df['Rol de la persona entrevistada']= df['Rol de la persona entrevistada'].str.replace('Dueño del negocio', '3')
df['Rol de la persona entrevistada']= df['Rol de la persona entrevistada'].str.replace('Empleado permanente', '2')
df['Rol de la persona entrevistada']= df['Rol de la persona entrevistada'].str.replace('Empleado temporal', '1')
df['Rol de la persona entrevistada']= df['Rol de la persona entrevistada'].str.replace('Familiar del propietario', '0')
```

```python
# Supongamos que 'Variable' es la columna que deseas modificar en el DataFrame 'df'
df['Tipo de negocio'] = df['Tipo de negocio'].apply(lambda x: 0 if df['Tipo de negocio'].value_counts()[x] < 6 else x)
```

```python
# Ver las frecuencias después de la modificación
frecuencias = df['Tipo de negocio'].value_counts()
print("Frecuencias después de la modificación:")
print(frecuencias)
```

```python
df['Tipo de negocio']= df['Tipo de negocio'].str.replace('Pequeño restaurante o fonda', '1')
df['Tipo de negocio']= df['Tipo de negocio'].str.replace('Tienda de alimentos/abarrotes', '2')
df['Tipo de negocio']= df['Tipo de negocio'].str.replace('Cafetería', '3')
df['Tipo de negocio']= df['Tipo de negocio'].str.replace('Tienda de tecnología/servicios tecnológicos', '4')
df['Tipo de negocio']= df['Tipo de negocio'].str.replace('Papelería', '5')
df['Tipo de negocio']= df['Tipo de negocio'].str.replace('Panadería', '6')
df['Tipo de negocio']= df['Tipo de negocio'].str.replace('Carnicería', '7')
df['Tipo de negocio']= df['Tipo de negocio'].str.replace('Tienda de ropa', '8')
df['Tipo de negocio']= df['Tipo de negocio'].str.replace('Lavandería', '9')
df['Tipo de negocio']= df['Tipo de negocio'].str.replace('Heladería', '10')
df['Tipo de negocio']= df['Tipo de negocio'].str.replace('Dulcería', '11')
df['Tipo de negocio']= df['Tipo de negocio'].str.replace('Estética', '12')
df['Tipo de negocio']= df['Tipo de negocio'].str.replace('Boutique', '13')
df['Tipo de negocio']= df['Tipo de negocio'].str.replace('Florería', '14')
df['Tipo de negocio']= df['Tipo de negocio'].str.replace('Veterinaria', '15')
df['Tipo de negocio']= df['Tipo de negocio'].str.replace('Pastelería', '16')
```

```python
#Ver las frecuencias después de la modificación
frecuencias = df['Tipo de negocio'].value_counts()
print("Frecuencias después de la modificación:")
print(frecuencias)
```

```python
#Sustituir valores nulos por un valor numérico en concreto
df["Tipo de negocio"] = df["Tipo de negocio"].fillna(0)
```

```python
#Conversión de tipo de dato de columna de tipo Object a Float
df['Tipo de negocio']= df['Tipo de negocio'].astype(int)
```

```python
# Obtener el nombre de la columna en la posición 0
columna_a_eliminar = df.columns[0]
columna_a_eliminar
```

```python
# Eliminar la columna en la posición 0
df = df.drop(columna_a_eliminar, axis=1)


df
```

```python
model.fit(X=Vars_Indep, y=Var_Dep)
```
